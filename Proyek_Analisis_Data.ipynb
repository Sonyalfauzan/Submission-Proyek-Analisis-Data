## Visualization & Explanatory Analysis

Pertanyaan 1: Bagaimana pengaruh cuaca terhadap jumlah penyewaan sepeda?

```python
# Membuat barplot rata-rata jumlah penyewaan berdasarkan jenis cuaca
plt.figure(figsize=(10, 6))
sns.barplot(x='weathersit', y='cnt', data=hour_df)
plt.title('Rata-rata Jumlah Penyewaan Sepeda Berdasarkan Jenis Cuaca')
plt.xlabel('Jenis Cuaca (1: Cerah, 2: Berkabut, 3: Hujan Ringan, 4: Hujan Lebat)')
plt.ylabel('Jumlah Penyewaan')
plt.show()

# Menghitung korelasi antara cnt dan weathersit
correlation = hour_df['cnt'].corr(hour_df['weathersit'])
print("Korelasi antara cnt dan weathersit:", correlation)

# Insight
print("\nInsight:")
print("- Terlihat bahwa jumlah penyewaan sepeda tertinggi terjadi saat cuaca cerah (weathersit 1).")
print("- Seiring dengan memburuknya cuaca, rata-rata jumlah penyewaan sepeda menurun.")
print("- Korelasi negatif (-0.297391) menandakan hubungan terbalik, artinya semakin buruk cuaca, semakin sedikit penyewaan.")
```

Pertanyaan 2: Bagaimana pola penyewaan sepeda berdasarkan hari dalam seminggu dan jam dalam sehari?

```python
# Membuat heatmap jumlah penyewaan berdasarkan hari dan jam
weekday_hour_df = hour_df.groupby(['weekday', 'hr'])['cnt'].mean().reset_index()
weekday_hour_df = weekday_hour_df.pivot(index='weekday', columns='hr', values='cnt') # Use keyword arguments for index, columns, and values
plt.figure(figsize=(12, 6))
sns.heatmap(weekday_hour_df, cmap='coolwarm')
plt.title('Pola Penyewaan Sepeda Berdasarkan Hari dan Jam')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Hari dalam Seminggu (0: Minggu, 6: Sabtu)')
plt.show()

# Membuat lineplot jumlah penyewaan berdasarkan hari, dibedakan antara hari kerja dan akhir pekan
plt.figure(figsize=(10, 6))
sns.lineplot(x='hr', y='cnt', hue='workingday', data=hour_df)
plt.title('Pola Penyewaan Sepeda Berdasarkan Jam di Hari Kerja dan Akhir Pekan')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Jumlah Penyewaan')
plt.show()

# Insight
print("\nInsight:")
print("- Terlihat lonjakan penyewaan pada jam sibuk (7-9 pagi dan 17-19 sore) di hari kerja, menunjukkan pola penggunaan untuk pergi dan pulang kerja.")
print("- Pada akhir pekan, puncak penyewaan terjadi di siang hari (10-16), menunjukkan pola penggunaan untuk rekreasi.")
```

## Analisis Lanjutan (Opsional)

RFM Analysis


```python
# Menghitung Recency, Frequency, dan Monetary
rfm_df = day_df.groupby('dteday')['registered'].sum().reset_index()
rfm_df['dteday'] = pd.to_datetime(rfm_df['dteday'])
recent_date = rfm_df['dteday'].max()
rfm_df['Recency'] = (recent_date - rfm_df['dteday']).dt.days
rfm_df['Frequency'] = 1
rfm_df['Monetary'] = rfm_df['registered']

# Melakukan segmentasi pelanggan (contoh sederhana)
rfm_df['RFM_Score'] = rfm_df['Recency'].astype(str) + rfm_df['Frequency'].astype(str) + rfm_df['Monetary'].astype(str)
rfm_df['Segment'] = 'Low-Value'
rfm_df.loc[rfm_df['RFM_Score'].str.startswith('01'), 'Segment'] = 'High-Value'

# Membuat scatterplot segmentasi pelanggan berdasarkan Recency dan Monetary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Recency', y='Monetary', hue='Segment', data=rfm_df)
plt.title('Segmentasi Pelanggan Berdasarkan RFM Analysis')
plt.show()

# Insight
print("\nInsight:")
print("- Pelanggan dengan nilai Recency rendah (baru-baru ini menyewa) dan Monetary tinggi (total penyewaan tinggi) termasuk dalam segmen High-Value.")
print("- Anda dapat menerapkan logika segmentasi yang lebih kompleks berdasarkan kuartil atau clustering.")
```

Clustering

```python
# Melakukan binning pada jam dalam sehari
hour_df['hour_group'] = pd.cut(hour_df['hr'], bins=[0, 6, 12, 18, 24], labels=['Dini Hari', 'Pagi', 'Siang', 'Malam'], right=False)

# Membuat barplot jumlah penyewaan berdasarkan kelompok jam
plt.figure(figsize=(10, 6))
sns.barplot(x='hour_group', y='cnt', data=hour_df)
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Kelompok Jam')
plt.xlabel('Kelompok Jam')
plt.ylabel('Jumlah Penyewaan')
plt.show()

# Insight
print("\nInsight:")
print("- Jumlah penyewaan tertinggi terjadi pada jam sibuk pagi (7-9) dan sore (17-19).")
print("- Anda dapat mengelompokkan data berdasarkan kriteria lain, misal: weekday dan workingday, untuk analisis lebih lanjut.")
```
